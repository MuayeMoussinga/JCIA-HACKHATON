```python
import os
import os.path as op
import shutil
import cv2
import numpy as np
import shutil
```

```python
#fonction de tri en fonction du floue
def is_blur(img,img_path,seuil):
    laplacien=cv2.Laplacian(img,cv2.CV_64F)
    variance=laplacien.var()
    print(f"L'image '{img_path}' a une variance de {variance}.")
    if (variance<seuil):
        os.remove(img_path)
        print(f"L'image '{img_path}' a été supprimée pour cause de flou.")
    
#fonction de redimensionnement
def resize(img,file,temporary_file,img_name,target_size=(640, 640)):
    resized_img = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)
    os.mkdir(temporary_file)
    original_path=os.path.join(file,img_name)
    new_path=os.path.join(temporary_file,img_name)
    cv2.imwrite(new_path, resized_img)
    shutil.copy(new_path,original_path)
    shutil.rmtree(temporary_file)
    print(f"L'image '{img_name}' a été redimensionnée et remplacée avec succès.")

#fonctioin pour supprimer les doublons
def remove_duplicates_xor(folder_path,image_dict=None):
    if image_dict is None:
        images ={}
    else:
        images=image_dict.copy()
    
    print(f"FICHIER : {folder_path}")
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if os.path.isfile(file_path):
            img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                for saved_path, saved_img in images.items():
                    diff = cv2.bitwise_xor(img, saved_img)
                    if not cv2.countNonZero(diff):  # Si tous les pixels sont identiques (XOR = 0 partout)
                        os.remove(file_path)
                        print(f"{filename} supprimée (doublon détecté par XOR)")
                        break
                    
                images[file_path] = img

#fonction pour déplacer les fichier
def move_images(file):
    file_path_and_name={
        "bruised":"C:\\Users\\User\\Desktop\\HACKATON2025\\african_plums\\bruised",
        "cracked":"C:\\Users\\User\\Desktop\\HACKATON2025\\african_plums\\cracked",
        "rotten":"C:\\Users\\User\\Desktop\\HACKATON2025\\african_plums\\rotten",
        "spotted":"C:\\Users\\User\\Desktop\\HACKATON2025\\african_plums\\spotted",
        "unaffected":"C:\\Users\\User\\Desktop\\HACKATON2025\\african_plums\\unaffected",
        "unripe":"C:\\Users\\User\\Desktop\\HACKATON2025\\african_plums\\unripe"
    }
    for name_images in os.listdir(file):
        for key in file_path_and_name.keys():
            if key.lower() in name_images.lower():
                file_name=os.path.join(file,name_images)
                shutil.move(file_name,file_path_and_name[key])
                print(f"Déplacement de '{file_name}' vers '{file_path_and_name[key]}'")
                break
                
```
